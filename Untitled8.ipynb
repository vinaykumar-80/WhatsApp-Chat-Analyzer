```python
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
print("Uploaded:", "WhatsApp Chat with Benhur Harrison.txt")
```

```python
import pandas as pd
import re

# Read file safely
with open(file_name, 'r', encoding='utf-8') as f:
    data = f.read()

# Regex pattern for WhatsApp format
pattern = r'(\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?[APMapm]{0,2})\s-\s'

messages = re.split(pattern, data)[1:]

dates = messages[::2]
msgs = messages[1::2]

df = pd.DataFrame({'date': dates, 'message': msgs})

# Extract user safely
users = []
texts = []

for msg in df['message']:
    if ': ' in msg:
        user, text = msg.split(': ', 1)
    else:
        user = "group_notification"
        text = msg
    users.append(user)
    texts.append(text)

df['user'] = users
df['text'] = texts

# Remove system messages
df = df[df['user'] != "group_notification"]

print("Data Loaded Successfully")
print("Total Messages:", len(df))
print("Users:", df['user'].unique())
df.head()
```

```python
import matplotlib.pyplot as plt

# Messages per user
df['user'].value_counts().plot(kind='bar')
plt.title("Messages per User")
plt.xlabel("User")
plt.ylabel("Count")
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(df['text'])

wc = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.imshow(wc)
plt.axis('off')
plt.title("WordCloud")
plt.show()
```

```python
import emoji
from collections import Counter

def extract_emojis(s):
    return [c for c in s if c in emoji.EMOJI_DATA]

emojis = []
for msg in df['text']:
    emojis.extend(extract_emojis(msg))

emoji_counts = Counter(emojis)

print("Top Emojis:")
print(emoji_counts.most_common(10))
```

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

df['sentiment'] = df['text'].apply(lambda x: sia.polarity_scores(str(x))['compound'])

print("Average Sentiment:", df['sentiment'].mean())
```

```python
# Convert date safely
df['date_parsed'] = pd.to_datetime(df['date'], errors='coerce')

df['hour'] = df['date_parsed'].dt.hour

df['hour'].value_counts().sort_index().plot(kind='bar')
plt.title("Activity by Hour")
plt.xlabel("Hour")
plt.ylabel("Messages")
plt.show()
```

```python
from transformers import pipeline

# Load sentiment analysis pipeline (uses a pretrained transformer model)
sentiment_model = pipeline("sentiment-analysis")

# Take a small sample (to keep it fast in Colab)
sample_texts = df['text'].dropna().astype(str).head(20).tolist()

# Run prediction
results = sentiment_model(sample_texts)

# Show results
for txt, res in zip(sample_texts, results):
    print("Text:", txt[:80])
    print("Sentiment:", res['label'], "| Confidence:", round(res['score'], 3))
    print("-"*50)
```
